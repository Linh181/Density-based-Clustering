## Generated data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import cluster, datasets
```

```python
from sklearn.datasets import make_blobs

n_samples = 10000

X, y = make_blobs(n_samples=n_samples, 
                  n_features=2, 
                  centers=4,
                  cluster_std=0.5,
                  random_state=1000)

fig=plt.figure(figsize=(8,8), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(X[:, 0], X[:, 1]);
#plt.savefig('blob.png');
```

```python
n_samples = 10000

X, y = make_blobs(n_samples=n_samples, 
                  n_features=2, 
                  centers=3,
                  cluster_std=0.4,
                  random_state=1000)
transformation = [[0.8, -0.8], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)
aniso = (X_aniso, y)

fig=plt.figure(figsize=(8,8), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(X_aniso[:, 0], X_aniso[:, 1]);
#plt.savefig('aniso.png');
```

```python
from sklearn.datasets import make_circles

n_samples = 10000

X, y = make_circles(n_samples=n_samples, 
                    factor=0.3,
                    noise=0.03,
                    random_state=1000)

fig=plt.figure(figsize=(8,8), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(X[:, 0], X[:, 1]);
#plt.savefig('circle.png');
```

```python
from sklearn.datasets import make_moons

n_samples = 10000

X, y = make_moons(n_samples=n_samples,
                  noise=0.01,
                  random_state=1000)

fig=plt.figure(figsize=(8,8), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(X[:, 0], X[:, 1]);
#plt.savefig('moon.png');
```

In this notebook, the efficiency of the two DBSCAN clustering algorithms using QuadTree and BoxGraph is investigated under different data sets. By varying the number of observations and keeping other parameters fixed, the running time of both algorithms on initialization and clustering is analyzed. The results indicate that BoxGraph takes almost constant time to initialize and cluster the data, while QuadTree follows linear time trends. In all cases, the BoxGraph algorithm is always more efficient than the QuadTree algorithm.
